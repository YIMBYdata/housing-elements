```python
import pandas as pd
```

```python
results_25_ft = pd.read_csv('../results/apn_or_geo_matching_25ft_results.csv')
results_100_ft = pd.read_csv('../results/apn_or_geo_matching_100ft_results.csv')
pdevs = ['City', 'P(dev) for inventory', 'P(dev) for vacant sites', 'P(dev) for nonvacant sites']
pdevs_25_ft = results_25_ft[pdevs]
pdevs_100_ft = results_100_ft[pdevs]
pdevs_100_ft.columns = [c + " (100 ft buffer)" for c in pdevs_100_ft.columns]
```

```python
table_a1 = pd.concat((pdevs_25_ft, pdevs_100_ft), axis=1)
table_a1.drop('City (100 ft buffer)', axis=1, inplace=True)
```

```python
table_a1.set_index('City', inplace=True)
```

```python
table_a1 = round(table_a1 * 100, 1)
```

```python
table_a1 = table_a1.applymap(lambda x : str(x) + '%' if x == x else 'N/A')
```

```python
table_a1 = table_a1.reindex(sorted(table_a1.columns), axis=1)
```
